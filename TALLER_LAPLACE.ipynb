### Ecuacin diferencial

La ecuacin diferencia del sistema mecnico se obtiene al considerar el equilibrio de fuerzas ejercidas sobre la masa:

\begin{equation}
F_\text{S}(t) + F_\text{F}(t) + F_\text{I}(t) = F_\text{E}(t)
\end{equation}

Donde $F_\text{E}(t) $ denota una fuerza externa que acta sobre la masa. La fuerza $F_\text{S}(t)$ inducida por el resorte est dada por la [ley de Hooke](https://en.wikipedia.org/wiki/Hooke%27s_law)

\begin{equation}
F_\text{S}(t) = k y(t)
\end{equation}

Es comn modelar la fuerza de friccin $F_\text{F}(t)$ inducida por el amortiguador de manera proporcional a la velocidad de la masa

\begin{equation}
F_\text{F}(t) = c \frac{d y(t)}{dt}
\end{equation}

La fuerza inercial $F_\text{I}(t)$ debida a la aceleracin de la masa se da como

\begin{equation}
F_\text{I}(t) = m \frac{d^2 y(t)}{dt^2}
\end{equation}

Al introducir las fuerzas anteriormente halladas en el equilibrio de fuerzas se obtiene la ecuacin diferencial que describe el desplazamiento del pndulo elstico amortiguado

\begin{equation}
m \frac{d^2 y(t)}{dt^2} + c \frac{d y(t)}{dt} + k y(t) = F_\text{E}(t)
\end{equation}

Como consecuencia de la fuerza externa.

La ecuacin anterior constituye una EDO con coeficientes constantes. Puede interpretarse como un SLTI con la fuerza externa $x(t)=F_\text{E}(t)$ como seal de entrada y el desplazamiento de la masa como la seal de salida $y(t)$.

### **Aplicacin de la Transformada de Laplace**
Aplicamos la transformada de Laplace a ambos lados de la ecuacin, asumiendo condiciones iniciales cero:

$$
\mathcal{L} \{ m \ddot{y}(t) + c \dot{y}(t) + k y(t) \} = \mathcal{L} \{ F_E(t) \}
$$

Usamos la propiedad de la transformada de derivadas:

$$
\mathcal{L} \{ \ddot{y}(t) \} = s^2 Y(s), \quad \mathcal{L} \{ \dot{y}(t) \} = s Y(s)
$$

Entonces, la ecuacin en el dominio de Laplace queda:

$$
m s^2 Y(s) + c s Y(s) + k Y(s) = F_E(s)
$$

Factorizamos \( Y(s) \):

$$
Y(s) \left( m s^2 + c s + k \right) = F_E(s)
$$


### **Funcin de Transferencia**
La funcin de transferencia se define como:

$$
H(s) = \frac{Y(s)}{F_E(s)}
$$

Sustituyendo la ecuacin anterior:

$$
H(s) = \frac{1}{m s^2 + c s + k}
$$

## **Anlisis del circuito**

$i_1(t)$ en la malla izquierda, que circula por $L$ y $C$.

$i_2(t)$ en la malla derecha, que circula por $C$ y $R$.

La corriente en el capacitor es:

$$ i_C = i_1 - i_2 $$


### **Aplicamos Kirchhoff**
Aplicamos la Ley de Voltajes de Kirchhoff (LVK) en cada malla.

#### **Malla 1 (Izquierda)**
$$ V_i(s) = Z_L(s) I_1(s) + Z_C(s) (I_1(s) - I_2(s)) $$

Sustituyendo las impedancias:

- $Z_L(s) = sL$ (inductor)
- $Z_C(s) = \frac{1}{sC}$ (capacitor)

$$ V_i(s) = sL I_1(s) + \frac{1}{sC} (I_1(s) - I_2(s)) $$

Factorizando:

$$ V_i(s) = \left( sL + \frac{1}{sC} \right) I_1(s) - \frac{1}{sC} I_2(s) $$

#### **Malla 2 (Derecha)**
$$ 0 = Z_C(s) (I_2(s) - I_1(s)) + Z_R(s) I_2(s) $$

Sustituyendo las impedancias:

- $Z_R(s) = R$

$$ 0 = \frac{1}{sC} (I_2(s) - I_1(s)) + R I_2(s) $$

$$ \frac{1}{sC} I_2(s) - \frac{1}{sC} I_1(s) + R I_2(s) = 0 $$

Factorizando $I_2(s)$:

$$ \left( R + \frac{1}{sC} \right) I_2(s) = \frac{1}{sC} I_1(s) $$

$$ I_2(s) = \frac{\frac{1}{sC}}{R + \frac{1}{sC}} I_1(s) $$


### **Funcin de Transferencia $H(s)$**
La salida es el voltaje en la resistencia:

$$ V_o(s) = R I_2(s) $$

Sustituyendo $I_2(s)$:

$$ V_o(s) = R \cdot \frac{\frac{1}{sC}}{R + \frac{1}{sC}} I_1(s) $$

De la ecuacin de la malla 1, despejamos $I_1(s)$:

$$ I_1(s) = \frac{V_i(s) + \frac{1}{sC} I_2(s)}{sL + \frac{1}{sC}} $$

Reemplazamos $I_2(s)$:

$$ I_1(s) = \frac{V_i(s) + \frac{1}{sC} \cdot \frac{\frac{1}{sC}}{R + \frac{1}{sC}} I_1(s)}{sL + \frac{1}{sC}} $$

Factorizamos $I_1(s)$:

$$ I_1(s) \left( sL + \frac{1}{sC} - \frac{1}{sC} \cdot \frac{\frac{1}{sC}}{R + \frac{1}{sC}} \right) = V_i(s) $$

Resolviendo esta ecuacin, encontramos la funcion de transferencia

$$ H(s) = \frac{V_o(s)}{V_i(s)} = \frac{R s}{L s^2 + R s + \frac{1}{C}} $$

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl


# Definimos valores para cada caso
cases = {
    "Subamortiguado": (1, 2, 0.5),   #  < 1
    "Sobreamortiguado": (1, 2, 3),   #  > 1
    "Amortiguamiento crtico": (1, 2, 2)  #  = 1
}

for case, (m, k, c) in cases.items():
    wn = np.sqrt(k / m)  # Frecuencia natural
    zeta = c / (2 * np.sqrt(m * k))  # Factor de amortiguamiento
    wd = wn * np.sqrt(1 - zeta ** 2) if zeta < 1 else 0  # Frecuencia amortiguada

    # Funcin de transferencia en lazo abierto
    num = [1]
    den = [m, c, k]
    sys = ctrl.TransferFunction(num, den)

    # Anlisis temporal
    t, y_impulse = ctrl.impulse_response(sys)
    t, y_step = ctrl.step_response(sys)

    # Grficos
    plt.figure(figsize=(10, 6))

    plt.subplot(2, 2, 1)
    plt.title(f"{case} - Respuesta Impulso")
    plt.plot(t, y_impulse, label="Impulso")
    plt.grid()

    plt.subplot(2, 2, 2)
    plt.title(f"{case} - Respuesta Escaln")
    plt.plot(t, y_step, label="Escaln")
    plt.grid()

    plt.subplot(2, 2, 3)
    poles, zeros = ctrl.pzmap(sys, plot=True)

    plt.subplot(2,2,4)
    mag, phase, omega = ctrl.bode(sys, plot=False)
    plt.title(f"{case} - Diagrama de Bode")
    plt.semilogx(omega, 20*np.log10(mag), label="Magnitud (dB)")
    plt.semilogx(omega, phase, label="Fase ()")
    plt.grid()
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f"{case}:")
    print(f"  Frecuencia natural no amortiguada (wn): {wn:.2f} rad/s")
    print(f"  Factor de amortiguamiento (zeta): {zeta:.2f}")
    if zeta < 1:
        print(f"  Frecuencia amortiguada (wd): {wd:.2f} rad/s")
```

```python
import numpy as np
import control as ctrl
import matplotlib.pyplot as plt

# Definimos valores para cada caso
cases = {
    "Subamortiguado": (1, 2, 0.5),   #  < 1
    "Sobreamortiguado": (1, 2, 3),   #  > 1
    "Amortiguamiento crtico": (1, 2, 2)  #  = 1
}

for case, (m, k, c) in cases.items():
    wn = np.sqrt(k / m)  # Frecuencia natural
    zeta = c / (2 * np.sqrt(m * k))  # Factor de amortiguamiento
    wd = wn * np.sqrt(1 - zeta ** 2) if zeta < 1 else 0  # Frecuencia amortiguada

    # Funcin de transferencia en lazo abierto
    num = [1]
    den = [m, c, k]
    sys = ctrl.TransferFunction(num, den)

    # Funcin de transferencia en lazo cerrado con realimentacin unitaria
    sys_cl = ctrl.feedback(sys)

    # Anlisis temporal en lazo cerrado
    t, y_impulse = ctrl.impulse_response(sys_cl)
    t, y_step = ctrl.step_response(sys_cl)

    # Grficos
    plt.figure(figsize=(10, 6))

    plt.subplot(2, 2, 1)
    plt.title(f"{case} - Respuesta Impulso (Lazo Cerrado)")
    plt.plot(t, y_impulse, label="Impulso")
    plt.grid()

    plt.subplot(2, 2, 2)
    plt.title(f"{case} - Respuesta Escaln (Lazo Cerrado)")
    plt.plot(t, y_step, label="Escaln")
    plt.grid()

    plt.subplot(2, 2, 3)
    poles, zeros = ctrl.pzmap(sys_cl, plot=True)

    # Bode en lazo cerrado
    mag, phase, omega = ctrl.bode(sys_cl, plot=False)
    plt.subplot(2,2,4)
    plt.title(f"{case} - Diagrama de Bode (Lazo Cerrado)")
    plt.semilogx(omega, 20*np.log10(mag), label="Magnitud (dB)")
    plt.semilogx(omega, phase, label="Fase ()")
    plt.grid()
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f"{case}:")
    print(f"  Frecuencia natural no amortiguada (wn): {wn:.2f} rad/s")
    print(f"  Factor de amortiguamiento (zeta): {zeta:.2f}")
    if zeta < 1:
        print(f"  Frecuencia amortiguada (wd): {wd:.2f} rad/s")
```
